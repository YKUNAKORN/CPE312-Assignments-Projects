```python
from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
bitcoin_data = cg.get_coin_market_chart_by_id(id='bitcoin', vs_currency = 'usd',days=30)
bitcoin_data
```

```python
bitcoin_data['prices']
```

```python
import pandas as pd
data = pd.DataFrame(bitcoin_data['prices'], columns=['TimeStamp','Price'])
data.head(5)
```

```python
data['Date'] = pd.to_datetime(data['TimeStamp'], unit='ms')
data['Date'] 
```

```python
candlestick_data = data.groupby(data.Date.dt.date).agg({'Price':['min','max','first','last']})
candlestick_data
```

```python
import plotly.graph_objects as go
import plotly
plotlyfig = go.Figure(data=[go.Candlestick(x=candlestick_data.index,
                                    open = candlestick_data['Price']['first'],
                                    high = candlestick_data['Price']['max'],
                                    low = candlestick_data['Price']['min'],
                                    close = candlestick_data['Price']['last'])
                                  ])

plotlyfig.update_layout(xaxis_rangeslider_visible = False, xaxis_title = 'Date',
                  yaxis_title = 'Price (USD $)', title = 'Bitcoin Chart over 30 Days')
plotly.offline.plot(plotlyfig, filename = './bitcoin_candlestick_graph.html', auto_open = True)
```
